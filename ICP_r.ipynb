```r
target <- 0
ExpInd <- raw[,1] + 1
Y <- raw[,target+2]
preds <- c(2:ncol(raw))
preds <- preds[-(target+1)]
X <- data.matrix(raw[, preds])
X
```

```r
icp = ICP(X,Y,ExpInd,selection='all',alpha=0.001)
print(icp)
```

```r
icp$Coeff
```

```r
icp$acceptedSets
```

```r
# One-hot encode accepted sets
p <- ncol(X)
one_hot <- matrix(0, length(icp$acceptedSets), p)
i = 1
for (set in icp$acceptedSets) {    
    one_hot[i,set] <- 1
    i = i + 1
}
print(one_hot)
```

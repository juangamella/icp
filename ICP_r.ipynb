### Load data from python

```r
path = 'data/test_data.csv'
raw = read.csv(path, header=FALSE)
raw
```

```r
ncol(raw)
```

```r
target <- 2
ExpInd <- raw[,1] + 1
Y <- raw[,target+2]
preds <- c(2:ncol(raw))
preds <- preds[-(target+1)]
X <- data.matrix(raw[, preds])
```

```r
icp = ICP(X,Y,ExpInd,selection='all',alpha=0.001)
print(icp)
```

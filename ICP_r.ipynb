### Load data from python

```r
case_no = 4
path = 'causalicp/test/test_cases/'
filename = sprintf('%scase_%d.csv', path, case_no)
raw = read.csv(filename, header=FALSE)
raw
```

```r
ncol(raw)
```

```r
target <- 0
ExpInd <- raw[,1] + 1
Y <- raw[,target+2]
preds <- c(2:ncol(raw))
preds <- preds[-(target+1)]
X <- data.matrix(raw[, preds])
X
```

```r
icp = ICP(X,Y,ExpInd,selection='all',test='normal',alpha=0.001,gof=0.001)
print(icp)
```

```r
icp$acceptedSets
```

```r
icp$ConfInt
```

```r
icp$Coeff
```

```r
icp$pvalues
```

```r
Xint <- cbind(rep(1, nrow(X)), X)
Xint
```

```r
linm <- glm( Y ~ Xint[,c(1,2)] - 1, family= "gaussian", control=glm.control(maxit=10,epsilon=10^(-6)))
coef(linm)
```

```r
### Compute p-values as in R implementation
```

```r
uni <- unique(ExpInd)
K <- length(uni)
        ExpIndNEW <- list()
        for (uc in 1:K) {
            ExpIndNEW[[uc]] <- which(ExpInd == uni[uc])
            attr(ExpIndNEW[[uc]], "value") <- uni[uc]
        }
IN <- ExpIndNEW
```

```r
ki <- 1
x <- resid[IN[[ki]]]
y <- resid[-IN[[ki]]]
pval <- 2*min( t.test(x,y)$p.value, var.test(x,y)$p.value)
pval
```
